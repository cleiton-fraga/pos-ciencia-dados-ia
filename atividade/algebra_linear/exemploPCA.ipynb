# PCA (Principal Component Analysis) em Ciência de Dados

## O que é PCA?

A **Análise de Componentes Principais (PCA)** é uma técnica estatística de **redução de dimensionalidade**. Ela transforma um conjunto de variáveis possivelmente correlacionadas em um conjunto menor de variáveis não correlacionadas, chamadas **componentes principais**. Esses componentes são combinações lineares das variáveis originais e são ordenados de forma que o primeiro componente principal explica a maior parte da variância dos dados, o segundo explica a maior parte da variância remanescente, e assim por diante.

## Por que usar PCA?

- **Reduzir a dimensionalidade dos dados:** Em muitos problemas de ciência de dados, os conjuntos de dados possuem um grande número de variáveis (features), o que pode dificultar a análise, visualização e até mesmo o desempenho de algoritmos de machine learning. O PCA permite reduzir o número de variáveis mantendo a maior parte da informação relevante.
- **Facilitar visualização:** Dados com muitas dimensões são difíceis de visualizar. O PCA pode reduzir os dados para 2 ou 3 dimensões, tornando possível criar gráficos e identificar padrões, agrupamentos ou outliers.
- **Remover ruído e redundância:** Muitas vezes, algumas variáveis são altamente correlacionadas ou contêm apenas ruído. O PCA ajuda a eliminar essas redundâncias e ruídos, focando nos componentes que realmente carregam informação.
- **Melhorar desempenho de algoritmos:** Reduzindo a dimensionalidade, o PCA pode ajudar a diminuir o tempo de processamento e evitar o problema da “maldição da dimensionalidade”, melhorando a performance de modelos de machine learning.
- **Pré-processamento para outras técnicas:** O PCA é frequentemente utilizado como etapa de pré-processamento antes de aplicar algoritmos como clustering (ex: K-means) ou classificação, tornando os dados mais manejáveis e relevantes.

## Como funciona o PCA?

1. **Padronização dos dados:** É importante padronizar as variáveis para que todas tenham média zero e desvio padrão um, evitando que variáveis com escalas maiores dominem a análise.
2. **Cálculo da matriz de covariância:** Avalia como as variáveis estão relacionadas entre si.
3. **Cálculo dos autovetores e autovalores:** Os autovetores determinam a direção dos novos eixos (componentes principais) e os autovalores indicam a quantidade de variância explicada por cada componente.
4. **Seleção dos principais componentes:** Os componentes principais são ordenados pela quantidade de variância explicada. Normalmente, seleciona-se os primeiros componentes que juntos explicam uma porcentagem significativa da variância total (ex: 90% ou 95%).
5. **Transformação dos dados:** Os dados originais são projetados nos novos eixos formados pelos componentes principais selecionados.

## Limitações do PCA

- O PCA é uma técnica **linear**: não captura relações não-lineares entre as variáveis.
- Os componentes principais podem ser difíceis de interpretar, pois são combinações das variáveis originais.
- Sensível à escala das variáveis (por isso a padronização é fundamental).

```python
# Importando bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Carregando o dataset Iris (Dataset de exemplo)
iris = load_iris()
X = iris.data
y = iris.target

# Padronizando os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicando PCA para reduzir para 2 componentes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualizando os dados reduzidos
plt.figure(figsize=(8,6))
for target in np.unique(y):
    plt.scatter(X_pca[y==target, 0], X_pca[y==target, 1], label=iris.target_names[target])
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('PCA aplicado ao dataset Iris')
plt.legend()
plt.show()

# Explicando a variância explicada
print("Variância explicada por cada componente:", pca.explained_variance_ratio_)
print("Variância total explicada:", np.sum(pca.explained_variance_ratio_))
```
